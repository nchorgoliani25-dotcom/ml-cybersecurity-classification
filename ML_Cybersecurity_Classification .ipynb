# Machine Learning for Cybersecurity
## Linear vs Logistic Regression

### Objective
The goal of this notebook is to demonstrate the difference between
Linear and Logistic Regression and to apply Logistic Regression
to a cybersecurity classification problem.

### Theory Summary
Linear Regression is used to predict continuous numerical values,
while Logistic Regression is used for classification problems.
Logistic Regression outputs probabilities between 0 and 1, which
are mapped to classes.

In cybersecurity, Logistic Regression is commonly used for:
- Spam detection
- Phishing detection
- Intrusion Detection Systems (IDS)


### Dataset
We use a small labeled dataset with two features: **WordCount** and **LinkCount**.
The target variable **IsSpam** is set to **1 for spam** and **0 for not spam**.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
data = [
 [56,4,0],[53,1,0],[70,4,0],[60,4,0],[77,7,1],[72,0,0],
 [69,5,1],[64,3,0],[30,0,1],[48,6,1],[75,8,1],[55,3,0],
 [44,3,1],[40,5,1],[76,2,0],[55,5,1],[29,6,1],[56,9,1],
 [43,2,1],[54,6,1],[68,2,0],[54,1,0],[67,9,1],[55,3,0],
 [37,7,1],[68,8,1],[58,6,1],[51,0,0],[43,2,1],[42,8,1],
 [51,0,0],[56,8,1],[31,7,1],[68,0,0],[74,5,1],[32,4,1]
]

df = pd.DataFrame(data, columns=["WordCount","LinkCount","IsSpam"])
df.head()
```

```python
plt.figure(figsize=(6,5))
plt.scatter(df[df.IsSpam==0]["WordCount"], df[df.IsSpam==0]["LinkCount"], label="Not Spam")
plt.scatter(df[df.IsSpam==1]["WordCount"], df[df.IsSpam==1]["LinkCount"], label="Spam")
plt.xlabel("Word Count")
plt.ylabel("Link Count")
plt.legend()
plt.title("Spam Dataset Visualization")
plt.show()
```

```python
X = df[["WordCount","LinkCount"]]
y = df["IsSpam"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("Accuracy:", accuracy)
print("Confusion Matrix:")
print(conf_matrix)
```

## Conclusion
This notebook demonstrates how Logistic Regression can be applied
to classification problems in cybersecurity.
The model successfully classifies spam emails with good accuracy,
showing the practical usefulness of machine learning techniques
in security-related tasks.
